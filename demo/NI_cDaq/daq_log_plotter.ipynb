```python
import matplotlib.pyplot as plt
import numpy as np
import csv
import datetime
```

```python
filename = "daq_log.csv"
times = []
voltages = []
with open('daq_log.csv', mode='r', newline='') as log_file:
        csv_reader = csv.reader(log_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        for row in csv_reader:
            times.append(datetime.datetime.strptime(row[0], '%Y-%m-%d %H:%M:%S.%f'))
            voltages.append(float(row[1]))

plt.plot(times, voltages)
plt.xlabel('Time')
plt.ylabel('Power (V)')
```
