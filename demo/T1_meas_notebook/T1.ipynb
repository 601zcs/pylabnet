```python
import numpy as np
import csv
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python

filename = "D:\\Dropbox\\Dropbox (Lukin SiV)\\SiV Quick Sharing\\B16 python experiments\\histograms\\2021\\03\\31\\histogram_11_17_46.txt"

store = []

for row in csv.reader(open(filename), delimiter =" "):
    store.append(row)

X = np.array([float(i) for i in store[0]])
Y = np.array([float(i) for i in store[1]])
```

```python
def function(t, a, b, T1):
    return a - b*np.exp(-t/T1)
    
fit = curve_fit(function, np.array(X), np.array(Y), p0=[3000, 2100, 0.1])

plt.plot(X,Y)
plt.plot(X,function(X, fit[0][0], fit[0][1], fit[0][2]))

print("T1 estimate: " + str(fit[0][2]*1000)[:5] + "ms")
```

```python
import scipy 
counts = np.linspace(0, 20, 201)

poisson1 = np.zeros(len(counts))
poisson2 = np.zeros(len(counts))
for ii in range(len(counts)): 
    poisson1[ii] = np.exp(-10) * 10**counts[ii]/scipy.special.gamma(counts[ii])
    poisson2[ii] = np.exp(-1.5) * 1.5**counts[ii]/scipy.special.gamma(counts[ii])

plt.plot(counts, poisson1)
plt.plot(counts, poisson2)
```
