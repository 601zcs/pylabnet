```python
# Log client
from pylabnet.utils.logging.logger import LogClient
```

```python
# Hardware clients
from pylabnet.hardware.p_gen.ni_hsdio import ni654x_spgen
from pylabnet.hardware.cw_mw.rs import rs_smc
from pylabnet.hardware.counter.swabian_instruments import gated_ctr
```

```python
# ODMR script
from pylabnet.scripts.odmr import ODMR
# GUI
from pylabnet.gui.igui.iplot import ComboTraceHMapPBar
```

## Connect to devices

```python
p_gen = ni654x_spgen.Client(host='localhost', port=1)
```

```python
# p_gen.connect()
```

```python
mw_src = rs_smc.Client(host='localhost', port=2)
```

```python
# mw_src.connect()
```

```python
gated_ctr = gated_ctr.Client(host='localhost', port=3)
```

```python
# gated_ctr.connect()
```

## GUI

```python
gui = ComboTraceHMapPBar(trace_title='Average', hm_title='All sweeps')
```

## ODMR class

```python
my_odmr = ODMR()
```

### set_hardware()

```python
my_odmr.set_hardware(
    gated_ctr=gated_ctr, 
    mw_src=mw_src, 
    p_gen=p_gen
)
```

### set_gui()

```python
my_odmr.set_gui(
    trace=gui.trace_fig, 
    heat_map=gui.hm_fig, 
    p_bar=gui.p_bar
)
```

## Measurement

### set_params()

```python
my_odmr.set_params(
    start_f=2.8e9, 
    stop_f=2.95e9, 
    n_pts=250, 
    n_swps=1000, 
    mw_pwr=-15,  # -15
    accum_dur=0.01
)
```

### run()

```python
my_odmr.clear_data()
```

```python
my_odmr.run()
```

```python
gui.show()
```
